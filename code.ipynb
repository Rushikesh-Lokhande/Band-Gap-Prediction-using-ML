```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# ✅ Define features (X) and target (y)
features = ["density", "density_atomic", "formation_energy_per_atom", "energy_above_hull"]
target = "band_gap"
```

```python
# ✅ Remove missing values
df = df.dropna(subset=features + [target])
```

```python
# ✅ Split data into training & testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)
```

```python
y_train
```

```python
# ✅ Scale the data (important for some models)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train)
```

```python
# ✅ Make predictions
y_pred = model.predict(X_test_scaled)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# ✅ Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Model Performance:\nMSE: {mse:.4f}\nR² Score: {r2:.4f}")
```

```python
# ✅ Define a new material (Example)
new_material = pd.DataFrame({
    "density": [3.5],  
    "density_atomic": [12.0],  
    "formation_energy_per_atom": [-0.2],  
    "energy_above_hull": [0.05]
})
```

```python
# ✅ Scale the new material data
new_material_scaled = scaler.transform(new_material)

# ✅ Predict band gap
predicted_band_gap = model.predict(new_material_scaled)

print(f"Predicted Band Gap: {predicted_band_gap[0]:.4f} eV")
```
